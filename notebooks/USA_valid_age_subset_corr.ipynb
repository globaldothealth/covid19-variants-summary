```python
import pandas as pd
import matplotlib.pyplot as plt

USA_weekly = pd.read_csv("../output/USA_weekly.csv")
USA_weekly_valid_age = pd.read_csv("../output/USA_valid-age_weekly.csv")
df = USA_weekly.merge(USA_weekly_valid_age, on=["Country", "Week"])

def week_to_float(weekstr: str) -> float:
    w = str(weekstr)
    assert len(w) == 6
    year, week = w[:4], w[4:]
    return int(year) + (int(week)- 1) / 52

df["WeekF"] = df.Week.map(week_to_float)
```

```python
plt.figure()
plt.plot(df.WeekF, df.Total_x)
plt.plot(df.WeekF, df.Total_y)
plt.title("Overall")

plt.figure()
plt.plot(df.WeekF, df.Delta_x)
plt.plot(df.WeekF, df.Delta_y)
plt.title("Delta")

plt.figure()
plt.plot(df.WeekF, df.Omicron_x)
plt.plot(df.WeekF, df.Omicron_y)
plt.title("Omicron")
```

```python
print("Overall correlation", df.Total_x.corr(df.Total_y))
print("Omicron correlation", df.Omicron_x.corr(df.Omicron_y))
print("  Delta correlation", df.Delta_x.corr(df.Delta_y))
```
